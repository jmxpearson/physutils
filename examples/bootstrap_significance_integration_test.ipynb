```python
import numpy as np
import pandas as pd
import physutils as phys
import physutils.bootstrap as boot
import physutils.tf as tf
import matplotlib.pyplot as plt
%matplotlib inline
np.random.seed(12345)
```

```python
lfp = phys.LFPset(pd.DataFrame.from_csv("sample_lfp.csv")).censor()
```

```python
evt = pd.DataFrame.from_csv('sample_events.csv')
```

```python
stops = evt['stop inflating'].dropna()
pops = evt['popped'].dropna()
```

```python
Tpre = -1.5
Tpost = 0.5
```

```python
baseline_interval = (-1.5, -1.35)
```

```python
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
```

```python
series = lfp.dataframe['17']
```

```python
nf = phys.norm_by_mean(baseline_interval)
contr_tf, fig = lfp.contrast_time_frequency('17', [stops, pops], 
                                            Tpre, Tpost, 
                                            method='wav',
                                            normfun=None, 
                                            freqs=freqs,
                                            clim=(-3, 3),
                                            cmap='seismic')
```

```python
%%time
lothresh = 0.15
hithresh = 0.85
thresh = (lothresh, hithresh)
mcontr, fig = lfp.significant_time_frequency('17', [stops, pops], 
                                             Tpre, Tpost, thresh, 
                                             niter=100, method='wav', 
                                             doplot=True, 
                                             normfun=None,
                                             freqs=freqs, clim=(-3, 3),
                                             cmap='seismic')
```
