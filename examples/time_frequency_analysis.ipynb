# Example: Time-Frequency Analysis

Here, we'll illustrate some time-frequency analysis of local field potentials (LFP).

To begin, do the usual bookkeeping and imports.

```python
import numpy as np
import pandas as pd
import physutils as phys
import physutils.bootstrap as boot
import physutils.tf as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

Fetch some LFP data to play with:

```python
lfp = phys.LFPset(pd.DataFrame.from_csv("sample_lfp.csv"))
lfp.head()
```

Get some example events from a behavioral task, the Balloon Analogue Risk Task (BART):

```python
evt = pd.DataFrame.from_csv('sample_events.csv')
evt.head()
```

Let's compare responses to two sets of events, voluntary stops and unexpected balloon pops:

```python
stops = evt['stop inflating'].dropna()
pops = evt['popped'].dropna()
```

We'll look at a window that begins 1.5 seconds before the event and ends half a second after.

```python
Tpre = -1.5
Tpost = 0.5
```

Now plot a moving window spectrogram that averages across trials:

```python
window = 0.250
percent_overlap = 0.95
spec, fig = lfp.avg_time_frequency('17', stops, Tpre, Tpost, method='spec', winlen=window, 
                              frac_overlap=percent_overlap)
```

For contrast, try a wavelet-based time-frequency analysis:

```python
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
wav, fig = lfp.avg_time_frequency('17', stops, Tpre, Tpost, method='wav', freqs=freqs)
```

We can also specify a normalization function to use with `avg_time_frequency`. This is useful because we want to compare changes within each frequency band, which have very different baseline power levels. Builtin options are `norm_by_trial` and `norm_by_mean` (across trials), but both require a baseline window.

```python
baseline_interval = (-1.5, -1.35)
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
nf = phys.norm_by_mean(baseline_interval)
wav_normed, fig = lfp.avg_time_frequency('17', stops, Tpre, Tpost, 
                                         method='wav', 
                                         freqs = freqs,
                                         normfun=nf)
```

Let's try this for another event type:

```python
baseline_interval = (-1.5, -1.35)
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
nf = phys.norm_by_mean(baseline_interval)
wav_normed, fig = lfp.avg_time_frequency('17', pops, Tpre, Tpost, 
                                         method='wav', 
                                         freqs = freqs,
                                         normfun=nf)
```

Let's try a contrast plot between the two event types:

```python
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
nf = phys.norm_by_mean(baseline_interval)
contr_tf, fig = lfp.contrast_time_frequency('17', [stops, pops], 
                                            Tpre, Tpost, method='wav',
                                            normfun=nf,
                                            freqs=freqs)
```

But are any clusters in this image statistically significant? 

Our strategy will be to shuffle labels for the two trial types and take any clusters in the image above a threshold that exceed the size of 95% of those in the null distribution.

First, we make a dataframe containing relevant events and their labels:

```python
t0 = pd.DataFrame({'time':stops, 'label':0})
t1 = pd.DataFrame({'time':pops, 'label':1})
alltimes = pd.concat([t0, t1])
labels = alltimes['label']
d0 = sum(labels == 0)
d1 = sum(labels == 1)
alltimes.head()
```

Next, we get a time-frequency matrix for the time segment corresponding to each event:

```python
series = lfp.dataframe['17']
tffun = tf.continuous_wavelet
baseline_interval = (-1.5, -1.35)
freqs = np.exp(np.linspace(np.log(2.5), np.log(50)))
nf = phys.norm_by_mean(baseline_interval)
spec0, taxis, faxis = tf._per_event_time_frequency(series, tffun, alltimes['time'][labels == 0], Tpre, Tpost, freqs=freqs, normfun=nf)
spec1, taxis, faxis = tf._per_event_time_frequency(series, tffun, alltimes['time'][labels == 1], Tpre, Tpost, freqs=freqs, normfun=nf)
spec0 = nf(spec0)
spec1 = nf(spec1)
spectra = spec0 + spec1
spectra = [s.values for s in spectra]
```

We will need to set a threshold for this statistic. 

```python
lothresh = -1
hithresh = 1
```

```python
pos = boot.make_thresholded_diff(spectra, labels, hi=hithresh, diff_fun=boot.normalized_diff_mean_power)
neg = boot.make_thresholded_diff(spectra, labels, lo=lothresh, diff_fun=boot.normalized_diff_mean_power)
```

```python
def plot_diff_map(img, taxis, faxis):
    ax = plt.figure().add_subplot(111)
    extent = (taxis[0], taxis[-1], faxis[0], faxis[-1])
    im = tf.NonUniformImage(ax, interpolation='bilinear', extent=extent)
    im.set_data(taxis, faxis, img.T)
    ax.set_xlim(extent[0], extent[1])
    ax.set_ylim(extent[2], extent[3])
    ax.images.append(im)
    plt.colorbar(im)
```

```python
multarray = np.array(spectra) # convert to array along dim 0
lls = np.array(labels)  # make sure this is an array
# convert to log scale 
arr0 = np.log(multarray[lls == 0])
arr1 = np.log(multarray[lls == 1])

p0 = np.log(np.nanmean(multarray[lls == 0], axis=0))
p1 = np.log(np.nanmean(multarray[lls == 1], axis=0))

m0 = np.nanmean(arr0, axis=0)
m1 = np.nanmean(arr1, axis=0)

s0 = np.nanstd(arr0, axis=0)
s1 = np.nanstd(arr1, axis=0)

n0 = np.sum(lls == 0)
n1 = np.sum(lls == 1)

numer = m0 - m1 + 0.5 * (s0 ** 2 - s1 ** 2)
denom = np.sqrt((s0 ** 2 / n0) + (s1 **2 / n1) + (s0 ** 4 / (n0 - 1)) +
        (s1 ** 4 / (n1 - 1)))

wald = numer / denom
```

```python
plot_diff_map(numer, taxis, faxis)
plot_diff_map(denom, taxis, faxis)
plot_diff_map(s0, taxis, faxis)
```

Plot statistic map, along with positive and negative clusters:

```python
plot_diff_map(pos.data, taxis, faxis)
plot_diff_map(pos, taxis, faxis)
plot_diff_map(neg, taxis, faxis)
```

```python
posclus = boot.label_clusters(pos)
negclus = boot.label_clusters(neg)
plot_diff_map(posclus, taxis, faxis)
plot_diff_map(negclus, taxis, faxis)
```

```python
cluster_masses = np.concatenate([boot.get_cluster_masses(pos, posclus)[1:], 
                               boot.get_cluster_masses(neg, negclus)[1:]])
np.sort(cluster_masses)
```

Now, do the same thing a whole bunch of times:

```python
%%time
thresh = (lothresh, hithresh)
niter = 100
pval = 0.05

try: 
    thlo = thresh[0]
    thhi = thresh[1]
except:
    thlo = thresh
    thhi = thresh

cluster_masses = np.array([])
for ind in np.arange(niter):
    labels = np.random.permutation(alltimes['label'])
    pos = boot.make_thresholded_diff(spectra, labels, hi=thhi, diff_fun=boot.t_of_log)
    neg = boot.make_thresholded_diff(spectra, labels, lo=thlo, diff_fun=boot.t_of_log)

    posclus = boot.label_clusters(pos)
    negclus = boot.label_clusters(neg)

    cluster_masses = np.concatenate([
        cluster_masses, 
        boot.get_cluster_masses(pos, posclus)[1:], 
        boot.get_cluster_masses(neg, negclus)[1:]
        ])

cluster_masses = np.sort(cluster_masses)
plo = pval / 2.0
phi = 1 - plo
Nlo = np.floor(cluster_masses.size * plo)
Nhi = np.ceil(cluster_masses.size * phi)
Clo = cluster_masses[Nlo]
Chi = cluster_masses[Nhi]
plt.hist(cluster_masses, 50, normed=1);
plt.vlines([Clo, Chi], *plt.ylim(), colors='red', linestyles='dashed')
print Clo, Chi
```

Let's use this to plot only significant clusters in the real data:

```python
truelabels = alltimes['label']
signif = boot.threshold_clusters(spectra, truelabels, lo=thlo, hi=thhi, 
                       keeplo=Clo, keephi=Chi, diff_fun=boot.log_F_stat)
plot_diff_map(signif, taxis, faxis)
```

Now retrieve the original contrast image from the spectra (doing this way saves the cost of computing the spectra all over again):

```python
truelabels = alltimes['label'].values
img0 = tf._mean_from_events(np.array(spectra)[truelabels == 0], taxis, faxis)
img1 = tf._mean_from_events(np.array(spectra)[truelabels == 1], taxis, faxis)
contrast = img0 / img1
```

We'll using masking to eliminate all entries that are not significant:

```python
mcontrast = contrast.copy()

mcontrast = mcontrast.mask(signif.mask)
```

```python
tf.plot_time_frequency(mcontrast);
```

Let's try all the above in one go with a method in the LFPset class:

```python
%%time
nf = phys.norm_by_mean(baseline_interval)
contr_tf, fig = lfp.contrast_time_frequency('17', [stops, pops], 
                                            Tpre, Tpost, 
                                            method='wav',
                                            normfun=nf, 
                                            freqs=freqs)
mcontr, fig = lfp.significant_time_frequency('17', [stops, pops], 
                                             Tpre, Tpost, thresh, 
                                             niter=100, method='wav', 
                                             doplot=True, 
                                             normfun=nf,
                                             freqs=freqs)
```
